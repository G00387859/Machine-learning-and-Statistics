Research:<br/> 
The operation of "a" to the power of a one half is the same as square root.<br/>
e.g. 3 to the power of 3 will equal 9. equivalence to 3 multiplied by 3 or 3 to the power of 2.<br/>
    9 to the power of one half equivalence to 9 to the power of (2*1/2) equals 9 <br/> 
     So ,the square root of 9 equivalence to 9 to the power of a 1/2. <br/>
Code expalined: The function sqrt2() below will take "a" a number as a parameter.<br/>
         The operation of "a" to the power of 1/2 is equivalence to the square root of "a"

```python
#[2]The variable "squareRoot" will be assigned the result of the equation.
squareRoot = 2**(1/2)
#[1]The variable formatedSqur will be assigned the formatted result.
formatedSqur="{:.100f}".format(squareRoot)
#print out the result
print(formatedSqur)
```
